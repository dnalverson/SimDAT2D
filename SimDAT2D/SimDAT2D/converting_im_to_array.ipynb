```python
import SimDAT2D as sim
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = sim.combine_image(sim.create_isotropic(.4, .4e-10, cmap = 'magma'), sim.create_anisotropic(15, 10, 200, size = 200, cmap = 'magma')*8)
```

```python
data_1d = data.reshape(-1)
```

```python
import numpy as np

# Generate 200 datasets of alterations
datasets = []
for _ in range(20):
    # Generate random values for multiplication, addition, subtraction, and division
    mult_factor = np.random.uniform(low=0.5, high=2.0)
    add_factor = np.random.uniform(low=-0.5, high=0.5)
    sub_factor = np.random.uniform(low=-0.5, high=0.5)
    div_factor = np.random.uniform(low=0.5, high=2.0)
    
    # Apply alterations to the data
    altered_data = (data_1d * mult_factor) + add_factor - sub_factor / div_factor
    
    # Append the altered dataset to the list
    datasets.append(altered_data)
```

```python
def get_clusters(data, number_clusters):
    # from sklearn.cluster import KMeans
    # kmeans = KMeans(n_clusters=number_clusters)
    # kmeans.fit(data)
    # return kmeans.predict(data

    from sklearn.cluster import AgglomerativeClustering
    Make_Clusters= AgglomerativeClustering(n_clusters = number_clusters, compute_distances=True)
    y_kmeans = Make_Clusters.fit_predict(data)
    information = Make_Clusters.fit(data)
    parameter = information.distances_

    x = 0
    Understanding_data = {"Cluster_Number":[], "Int_Angle":[]}
    while x < len(data):
        Understanding_data["Cluster_Number"].append(y_kmeans[x])
        Understanding_data["Int_Angle"].append(data[x])
        x = x + 1
        
    return Understanding_data
```
